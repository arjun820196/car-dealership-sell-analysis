Creating new column of mil_range with data type is numerical, by extracting data from mileage column.

```python
df_mileage=(data['Mileage'].str.extract(r'(?P<mil_range>\d+(?:\.\d+)?)').applymap(str.strip)).apply(pd.to_numeric)
```

Joining two dataframes to add new column (mil_range) to original dataset

```python
newdf=data.join(df_mileage)
```

Creating new column of Engine_size with data type is numerical, by extracting data from Engine column.

```python
newdf['Engine_size']=(data['Engine'].str.extract(r'(?P<Engine_size>\d+(?:\.\d+)?)').applymap(str.strip)).apply(pd.to_numeric)
```

Creating new column of Power_output with data type is numerical, by extracting data from Power column.

```python
# newdf['Power_output']=(data['Power'].str.extract(r'(?P<Power_output>\d+(?:\.\d+)?)').applymap(str.strip)).apply(pd.to_numeric)
```

```python
newdf['Power'].unique()
```

```python
newdf['Power_output']=newdf['Power'].str.split(n=1).str[0]
#n = int, default -1 (all):- Limit number of splits in output. None, 0 and -1 will be interpreted as return all splits.
```

```python
newdf['Power_output'] = newdf['Power_output'].replace(['null'], '74')  # 74 - is mode of Power output column
newdf = newdf.astype({"Power_output": float})
```

Creating new column of Brand in dataframe, by extracting data from Name column.

```python
newdf['Brand']=newdf['Name'].str.split(n=1).str[0] 
#n = int, default -1 (all):- Limit number of splits in output. None, 0 and -1 will be interpreted as return all splits.
```

Finding correlation between Price and Mileage of car

```python
'''
The correlation coefficient shows the direction and strength of a relationship between two variables.
The closer the correlation coefficient value is to +1 or -1, the stronger the linear relationship between the two variables is.
A coefficient of 1 shows a perfect positive correlation, or a direct relationship.
while 0 means there is no linear relationship.
'''
```

```python
print(newdf['mil_range'].corr(newdf['Price']))  # Correlation 
```

```python
# correlation coefficient for mileage and price of resold car is negative & very close to the zero. i.e -0.280
# Which indicates there is no stronge relation between mileage and price of the resold car.
# Mileage of any car does not affect the price of resold cars. 
```

Regression Model

---

```python
from sklearn.model_selection import train_test_split
```

```python
x=df['Power_output'].values.reshape(-1,1)
y=df['Price'].values.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)  # Model training
```

```python
reg.intercept_      # c = intercept
```

```python
reg.coef_        # m = slope
```

```python
y_pred=reg.predict(x_test)
```

```python
y_pred
```

```python
df_pred=pd.DataFrame({'Power_output':x_test.squeeze(),'Price':y_test.squeeze(), 'Predicted':y_pred.squeeze()})
```

```python
df_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mae=mean_absolute_error(y_test,y_pred)
```

```python
mae
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
rmse=np.sqrt(mse)
```

```python
rmse
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)
print(score)
```
